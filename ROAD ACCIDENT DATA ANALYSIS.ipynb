**The data set consists data from 1 Jan 2021 to 31 Dec 2020**

```python
d1=df['Year'].value_counts()
plt.figure(figsize=(2,2))
ax=sns.barplot(y=d1,x=d1.index,palette="tab10")
for bars in ax.containers:ax.bar_label(bars)
```

**In 2021 there were more number of accidents reported**

```python
d1=df['Urban_or_Rural_Area'].value_counts()
plt.figure(figsize=(2,2))
ax=sns.barplot(y=d1,x=d1.index,palette="rocket_r")
for bars in ax.containers:ax.bar_label(bars)
```

**In urban area there are more accidents**

**Accident severity on the basis of Speed**

```python
plt.figure(figsize=(4,4))
ax=sns.barplot(y='Speed_limit',x='Accident_Severity',hue='Urban_or_Rural_Area',data=df.groupby(['Accident_Severity','Urban_or_Rural_Area'],as_index=False)['Speed_limit'].mean(),palette="terrain_r")
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
for bars in ax.containers:ax.bar_label(bars)
```
